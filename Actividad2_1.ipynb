```python
data = pd.read_csv('Ventas_totales.csv')
data
```

```python
data2 = data.copy()
data2
```

```python
#Sustitucion por medio de la media la columna de salon_ventas
data2['salon_ventas'] = data2['salon_ventas'].fillna(round(data['salon_ventas'].mean(),13))
```

```python
#Sustitucion por medio de la media en la columna de tarjetas de debito
data2['tarjetas_debito'] = data2['tarjetas_debito'].fillna(round(data['tarjetas_debito'].mean(),12))
```

```python
#Sustitucion por medio de la media en la columna de tarjetas de debito
data2['tarjetas_credito'] = data2['tarjetas_credito'].fillna(round(data['tarjetas_credito'].mean(),12))
```

```python
#Sustitucion por medio de la media en la columna de otros_medios
data2['otros_medios'] = data2['otros_medios'].fillna(round(data['otros_medios'].mean(),3))
```

```python
#Sustituir con mediana la columna de subtotal_ventas_alimentos_bebidas
data2['subtotal_ventas_alimentos_bebidas'] = data2['subtotal_ventas_alimentos_bebidas'].fillna(round(data['subtotal_ventas_alimentos_bebidas'].median(),1))
```

```python
#Sustituir por valores no nulos hacia adelante "foward fill" (ffill) la columna de bebidas
data2['bebidas'] = data2['bebidas'].fillna(method="ffill")
```

```python
#Sustituir por valores no nulos hacia atr√°s "backward fill" (bfill)
data2['almacen'] = data2['almacen'].fillna(method="bfill")
```

```python
#Sustitucion por medio de la media en la columna de carnes
data2['carnes'] = data2['carnes'].fillna(round(data['carnes'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de panaderia
data2['panaderia'] = data2['panaderia'].fillna(round(data['panaderia'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de lacteos
data2['lacteos'] = data2['lacteos'].fillna(round(data['lacteos'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de verduleria_fruteria
data2['verduleria_fruteria'] = data2['verduleria_fruteria'].fillna(round(data['verduleria_fruteria'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de alimentos_preparados_rotiseria
data2['alimentos_preparados_rotiseria'] = data2['alimentos_preparados_rotiseria'].fillna(round(data['alimentos_preparados_rotiseria'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de indumentaria_calzado_textiles_hogar
data2['indumentaria_calzado_textiles_hogar'] = data2['indumentaria_calzado_textiles_hogar'].fillna(round(data['indumentaria_calzado_textiles_hogar'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de electronicos_articulos_hogar
data2['electronicos_articulos_hogar'] = data2['electronicos_articulos_hogar'].fillna(round(data['electronicos_articulos_hogar'].mean(),3))
```

```python
#Sustitucion por medio de la media en la columna de otros
data2['otros'] = data2['otros'].fillna(round(data['otros'].mean(),3))
```

Se decidio sustituir los valores nulos por la media y la mediana, en el caso de la media solo se cambiaron los decimales en algunas ya que se encontraban con notacion cientifica y para que no afecte se pusieron decimales de 12, en cambio con las otras columnas se usaron decimales de 3.
